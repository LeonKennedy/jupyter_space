#### 画图的使用

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-5, 5, 0.02)
y1 = np.sin(x)
```

+ figure, 作用新建绘画窗口,独立显示绘画的图片
+ subplot  多个图像画在同一个绘画窗口
 - r 行
 - c 列
 - n 第几个
+ xlim 和 ylim用来设置坐标范围

```python
plt.subplot(211)
plt.plot(x,y1)
plt.subplot(212)
from pylab import xlim, ylim
xlim(-2.5, 2.5)
ylim(-2,1)
plt.plot(x, y1)
```

#### 叠加图
+ 颜色

```python
x1 = np.arange(0,5,0.2)
plt.subplot(2,2,1)
plt.plot(x1,x1, 'r:', x1, x1**2, 'bo', x1,x1**3, 'g*')
plt.subplot(2,2,2)
plt.plot(x1,x1, 'm-.', x1, x1**2, 'y*', x1,x1**3, 'g+')
plt.subplot(2,2,3)
plt.plot(x1,x1, 'm_', x1, x1**2, 'kd', x1,x1**3, 'g8')
plt.show()
```

#### 背景色

```python
plt.subplot(111,facecolor=(0.1843,0.3098,0.3098))
plt.plot(x, x**2, 'c')
```

#### 可哦爹
+ plt.text 添加文字
+ plt.grid  添加网格
+ plt.hist  直方图

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
 
# 数据的直方图
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
 
 
plt.xlabel('Smarts')
plt.ylabel('Probability')
#添加标题
plt.title('Histogram of IQ')
#添加文字
plt.text(50, .025, r'$mu=100, sigma=15$')
plt.grid(True)
plt.show()
```
