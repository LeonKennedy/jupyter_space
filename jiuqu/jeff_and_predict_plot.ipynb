```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_pickle('202021.pkl')
```

```python
predict_mean = (data['teacher_count'] - data['predict']).abs().mean()
jeff_mean = (data['teacher_count'] - data['jeff']).abs().mean()
```

```python
plt.figure(figsize=(20, 8))
ax1 = plt.subplot(1, 1, 1)
ax1.plot(data.index.values, data['teacher_count'].values, label='teacher')
ax1.plot(data.index.values, data['predict'].values, label='predict')
ax1.plot(data.index.values, data['jeff'].values, label='jeff')
ax1.set_title(f"[week 202021] jeff predict MAE: {round(jeff_mean, 2) },     model predict MAE: {round(predict_mean, 2)}")
ax1.legend()
```
