## 到达满班率的时间

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('schedule_rate_full_time.pkl')
df['date'] = [s.time() for s in df.index]
```

```python
df['date'].unique()
```

```python
fig = plt.figure(0,figsize=(80, 60))
fig_index = 1
ax = fig.add_subplot(1,1,1)
for date, data in df.groupby('date'):
    ax = fig.add_subplot(3,4,fig_index)
    for room_type, sub_data in data[['room4', 'room2', 'room1']].items():
        ax.plot(sub_data.index, sub_data.values /24 , '--*', label=room_type)
    ax.set_title(str(date))
    ax.legend()
    fig_index += 1
plt.show()
```
