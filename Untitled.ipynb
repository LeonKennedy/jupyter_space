# 修订03

```python
bash_dir = "/tmp/pycharm_project_313"
teachers = pd.read_pickle(f"{bash_dir}/output/{week}/accident.pkl")
info = pd.read_pickle(f"{bash_dir}/output/{week}/teacher.pkl")
```

```python
c = teachers.appoint + teachers.standby > 0
teachers = teachers.loc[c]
```

```python
teachers.appoint.value_counts().sort_index().plot()
```

```python
num  = teachers.groupby('feedback').sum().iloc[:,0] /teachers.groupby('feedback').count().iloc[:,0]
```

```python
num.plot()
```

```python
teachers.sum()
```

```python
c = info.fresh == 1
fresh_teacher = info.loc[c]
c = fresh_teacher.index.intersection(teachers.index)
teachers.loc[c]
```
