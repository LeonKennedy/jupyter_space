```python
from connector import odps
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tablename = 'Qkids_V2.prt_student_consume_cohort'
table = odps.get_table(tablename)
temp_list = list()
for p in  table.partitions:
    for row in odps.read_table(tablename, partition = p.name):
        record = [*row.values, p.name[4:10]] 
        temp_list.append(record)
df = pd.DataFrame(temp_list, columns = ['registe', 'num', 'ds', 'month'])
```

```python
cohort = df.pivot_table(values= 'num', index='registe', columns= 'month')
cohort
```
